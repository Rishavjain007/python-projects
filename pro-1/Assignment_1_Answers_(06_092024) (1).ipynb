```python
# Answer 8
num = 5
factorial = 1
if num < 0:
    print("Factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    # run loop 5 times
    for i in range(1, num + 1):
        # multiply factorial by current number
        factorial = factorial * i
    print("The factorial of", num, "is", factorial)
```

```python
#Answer 9

num = 75962
reverse_number = 0
print("Given Number ", num)
while num > 0:
    reminder = num % 10
    reverse_number = (reverse_number * 10) + reminder
    num = num // 10
print("Revere Number ", reverse_number)
```

xplanation:
num = 75962

Here, a variable num is assigned the value 75962. This is the number you want to reverse.
reverse_number = 0

This initializes another variable reverse_number with 0. This variable will hold the reversed number as the loop progresses.
print("Given Number ", num)

This line simply prints the given number (75962) to the console.
while num > 0:

This starts a while loop that will continue executing as long as the value of num is greater than 0. The loop ensures that the digits are processed one by one from the last digit to the first.
reminder = num % 10

The % (modulo) operator returns the remainder when num is divided by 10. This operation extracts the last digit of num.
For example, 75962 % 10 = 2. So, reminder = 2 in the first iteration.
reverse_number = (reverse_number * 10) + reminder

This is the core logic of the reversal. It shifts the digits in reverse_number to the left (by multiplying by 10), then adds the extracted last digit (reminder).
First iteration:
reverse_number = (0 * 10) + 2 = 2
Second iteration:
reminder = 7596 % 10 = 6
reverse_number = (2 * 10) + 6 = 26
This process continues, building up the reversed number step by step.
num = num // 10

The // operator performs integer division (division without a remainder). This operation effectively removes the last digit of num.
First iteration:
num = 75962 // 10 = 7596
Second iteration:
num = 7596 // 10 = 759
This reduces the number with each iteration until it becomes 0, at which point the loop will stop.
print("Revere Number ", reverse_number)

Once the loop finishes, the final reversed number is printed.
Iteration Breakdown:
Iteration	num	reminder	reverse_number	Updated num#

1	         75962	2	2	7596
2	         7596	6	26	759
3	         759	9	269	75
4	         75	5	2695	7
5	         7	7	26957	0 (loop ends)
Final Output:
Given Number: 75962
Reversed Number: 26957
