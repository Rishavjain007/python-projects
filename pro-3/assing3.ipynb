The primary goal of Object-Oriented Programming (OOP) is to enhance code organization and reusability through the use of objects. OOP aims to model real-world entities and relationships, making it easier to design, maintain, and scale software systems.

Key principles of OOP include:

Encapsulation: Bundling data and methods that operate on that data within objects, which helps protect the data and maintain integrity.

Abstraction: Simplifying complex systems by modeling classes based on essential properties and behaviors, while hiding unnecessary details.

Inheritance: Allowing new classes to inherit properties and methods from existing classes, promoting code reuse and reducing redundancy.

Polymorphism: Enabling objects to be treated as instances of their parent class, allowing for flexible and interchangeable code.

Class Variables:

These are defined within a class 

They belong to the class itself rather than to any specific instance, means, if you change the value of a class variable, it will affect all instances of that class.

They are typically used to store properties that are common to all instances.

Instance Variables:

They are defined within methods (usually the __init__ method) using the self keyword and are specific to each instance of the class.

Each instance of the class has its own copy of instance variables which will not affect the instance variable in another object 
after making changes in one object

They are used to store data that is specific to each instance of the class.

It enables methods to know which instance they are working with, ensuring that operations are applied correctly to that specific object.

```python
#ans=7

class Book():
    def __init__(self,title,author,isbn,publication_year,available_copies):
        self.title=title
        self.author=author
        self.isbn=isbn
        self.publication_year=publication_year
        self.available_copies=available_copies

    def check_out(self):
        if self.available_copies > 0:
            self.available_copies -= 1
            print(f"check out of {self.title} is done")

        else:
            print(f"{self.title} is not available")

    def return_book(self):
         self.available_copies += 1
         print(f"{self.title} has been returned")

    def display_book_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"ISBN : {self.isbn}")
        print(f"Publication Year : {self.publication_year}")
        print(f"Available Copies : {self.available_copies}")

Book1=Book("The game of RJ","RJGAMER",85,2018,5)  

Book1.display_book_info() 

Book1.check_out()

Book1.display_book_info() 

Book1.return_book()

Book1.display_book_info()           
                    
```

```python
class Ticket():
    def __init__(self,ticket_id,event_name,event_data,venue,seat_number,price,is_reserved):
        self.ticked_id=ticket_id
        self.event_name=event_name
        self.event_data=event_data
        self.venue=venue
        self.seat_number=seat_number
        self.price=price
        self.is_reserved=is_reserved

    def reserve_ticket(self):
        if self.is_reserved != True:
            self.is_reserved = True
            print(f"{self.ticked_id} is now reserved")
        else:
            print(f"{self.ticked_id} is already reserved")

    def cancle_reservation(self):
        self.ticked_id == False
        print(f"reservation for {self.ticked_id} has been cancled")

    def display_ticket_info(self):
        print(f"Ticket ID :{self.ticked_id}")
        print(f"Event Name : {self.event_name}")
        print(f"Event Data : {self.event_data}")
        print(f"Venue : {self.venue}")
        print(f"Seat Number : {self.seat_number}")
        print(f"Price : {self.price}")
        print(f"Is Reserved : {self.is_reserved}")

ticket1=Ticket("Rishav","show","hello","RJ home",7,50,"True")                            
ticket1.display_ticket_info()
ticket1.reserve_ticket()
ticket1.display_ticket_info()
ticket1.cancle_reservation()
ticket1.display_ticket_info()
```
